# lãi suất, lạm phát

```python
df_g = pd.read_csv('/Users/user/gold_forecasting_timeseries/data/gold_5y_1d.csv')
df_g.head()
```

```python
# xoá các giá trị null:
df_g.dropna(inplace=True)
df_g
```

```python
# chuẩn hoá cột Date:
df_g['Date'] = pd.to_datetime(df_g['Date'])
```

#khối lượng giao dịch trên từng quý

```python
# tạo cột theo quý
df_g["year"] = df_g["Date"].dt.year
df_g["quarter_number"] = df_g["Date"].dt.quarter
```

```python
df_g["quarter"] = df_g["year"].astype(str) + "-Q" + df_g["quarter_number"].astype(str)
```

```python
# tính khối lượng giao dịch theo quý:
giaodich_quy = df_g.groupby('quarter')['Volume'].sum().reset_index()
giaodich_quy
```

```python
# tính giá trung bình của từng quý
avg_price_by_quarter = (
    df_g.groupby("quarter")["Close"]
    .mean()
    .reset_index()
    .rename(columns={"Close": "avg_close_price"})
)
avg_price_by_quarter
```

```python
# merge 2 data của giá trên từng quý và data tổng khối lượng giao dịch trên từng quý:
df_quarterly = pd.merge(giaodich_quy, avg_price_by_quarter, on="quarter")
df_quarterly
```

```python
avg_price_by_quarter
```

link đến trang lãi suất điều hành của FED dữ liệu lấy theo quý từ 2020 đến 2025:
https://fred.stlouisfed.org/series/FEDFUNDS#

```python
df_fed = pd.read_csv('FEDFUNDS (1).csv')
df_fed
```

```python
# tạo cột quý cho df_fed:
df_fed['observation_date'] = pd.to_datetime(df_fed['observation_date'])
df_fed['year'] = df_fed['observation_date'].dt.year
df_fed['quarter_number'] = df_fed['observation_date'].dt.quarter
```

```python
df_fed["quarter"] = df_fed["year"].astype(str) + "-Q" + df_fed["quarter_number"].astype(str)
```

```python
# gộp giá vàng và lãi suất theo guý
df_fed_gold = pd.merge(df_fed, avg_price_by_quarter, on="quarter")
df_fed_gold
```

```python
# lưu file:
df_fed_gold.to_csv('df_fed_gold.csv', index=False)
```
